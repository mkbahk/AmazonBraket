```python
city_names = {
    0: "서울", 1: "인천", 2: "경기(수원)", 3: "강원(춘천)",
    4: "충북(청주)", 5: "충남(홍성)", 6: "세종", 7: "대전",
    8: "전북(전주)", 9: "전남(무안)", 10: "경북(안동)", 11: "경남(창원)",
    12: "제주", 13: "부산", 14: "울산", 15: "대구", 16: "광주"
}

# 숫자로 된 경로를 지명으로 변환
city_permutation = [city_names[i] for i in permutation]

print(f"최적 경로 (숫자) 순서: {permutation}")
print(f"최적 경로 (지명) 순서: {city_permutation}")
print(f"최소 거리: {distance} Km")
```
