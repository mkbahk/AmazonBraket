
# RXX Gate
* RXX Gate는 양자 컴퓨팅에서 양자 얽힘(entanglement)을 생성하거나 조작하는 데 사용되는 two-qubit gate입니다.
* 이 게이트는 두 큐비트 상태를 Pauli-X 연산자의 텐서곱 X⊗X를 기준으로 특정 각도로 회전시특정 지점에서 큐비트 사이의 연산 순서를 고정하고, 회로 최적화 또는 재배치를 제한하는 데 사용




```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
import math

RXXGateCircuit = QuantumCircuit(2, 2)
RXXGateCircuit.rxx(math.pi/4, 0, 1)
```

```python
from qiskit.quantum_info import Operator
unitary = Operator(RXXGateCircuit)
print(unitary)
```

```python
from qiskit.visualization import array_to_latex

#Alternative way of representing in latex
array_to_latex(unitary)
```

```python
circuit = QuantumCircuit(2, 2)
circuit.rxx(math.pi, 0, 1)
circuit.measure([0, 1], [1, 1])
```

```python
circuit.draw(output='mpl')
```

```python
job = backend.run(circuit, shots=1024)
result = job.result()
```

```python
result
```

```python
result.results
```

```python
result.results[0].data.counts
```

```python
result.get_counts()
```

```python
counts = result.get_counts()
print(counts)
```

```python
from qiskit.visualization import plot_histogram, plot_state_city, plot_bloch_multivector,plot_state_paulivec, plot_state_hinton, plot_state_qsphere
plot_histogram(counts)
```
